## 1\. ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®ã‚¤ãƒ³ãƒãƒ¼ãƒˆ

ã¾ãšã€æ•°å€¤è¨ˆç®—ï¼ˆç‰¹ã«ãƒ™ã‚¯ãƒˆãƒ«ã‚„è¡Œåˆ—ã®è¨ˆç®—ï¼‰ã‚’ç°¡å˜ã«è¡Œã†ãŸã‚ã®ãƒ©ã‚¤ãƒ–ãƒ©ãƒª **NumPy** ã‚’ã‚¤ãƒ³ãƒãƒ¼ãƒˆã—ã¾ã™ã€‚

```python
import numpy as np
```

## 2\. é‡å­çŠ¶æ…‹ã®è¡¨ç¾ âš›ï¸

1é‡å­ãƒ“ãƒƒãƒˆã®çŠ¶æ…‹ã¯ã€2æ¬¡å…ƒã®è¤‡ç´ ãƒ™ã‚¯ãƒˆãƒ«ï¼ˆ**çŠ¶æ…‹ãƒ™ã‚¯ãƒˆãƒ«**ï¼‰ã§è¡¨ç¾ã•ã‚Œã¾ã™ã€‚

$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle = \begin{pmatrix} \alpha \\ \beta \end{pmatrix}$$

ã“ã“ã§ã€$\alpha$ ã¨ $\beta$ ã¯**ç¢ºç‡æŒ¯å¹…**ã¨å‘¼ã°ã‚Œã‚‹è¤‡ç´ æ•°ã§ã€ãã‚Œãã‚Œã®çµ¶å¯¾å€¤ã®2ä¹—ã®å’Œ ($|\alpha|^2 + |\beta|^2$) ã¯å¸¸ã«1ã«ãªã‚Šã¾ã™ã€‚

  - **$|0\rangle$ çŠ¶æ…‹**: è¨ˆç®—åŸºåº•çŠ¶æ…‹ã®ä¸€ã¤ã€‚ãƒ™ã‚¯ãƒˆãƒ«è¡¨ç¾ã¯ $\begin{pmatrix} 1 \\ 0 \end{pmatrix}$ ã§ã™ã€‚
  - **$|1\rangle$ çŠ¶æ…‹**: ã‚‚ã†ä¸€ã¤ã®è¨ˆç®—åŸºåº•çŠ¶æ…‹ã€‚ãƒ™ã‚¯ãƒˆãƒ«è¡¨ç¾ã¯ $\begin{pmatrix} 0 \\ 1 \end{pmatrix}$ ã§ã™ã€‚

ã“ã‚Œã‚’ã‚³ãƒ¼ãƒ‰ã§å®šç¾©ã—ã¾ã—ã‚‡ã†ã€‚

```python
# |0> çŠ¶æ…‹ãƒ™ã‚¯ãƒˆãƒ«
# np.arrayã§ãƒ™ã‚¯ãƒˆãƒ«ã‚’å®šç¾©ã—ã¾ã™ã€‚
# è¤‡ç´ æ•°ã§ã‚ã‚‹ã“ã¨ã‚’æ˜ç¤ºã™ã‚‹ãŸã‚ã« 'j' ã‚’ä½¿ã„ã¾ã™ã€‚
qbit_0 = np.array([1+0j, 0+0j])

# |1> çŠ¶æ…‹ãƒ™ã‚¯ãƒˆãƒ«
qbit_1 = np.array([0+0j, 1+0j])

print(f"|0> çŠ¶æ…‹: {qbit_0}")
print(f"|1> çŠ¶æ…‹: {qbit_1}")
```

## 3\. é‡å­ã‚²ãƒ¼ãƒˆã®è¡¨ç¾ ğŸšª

é‡å­ãƒ“ãƒƒãƒˆã®çŠ¶æ…‹ã‚’æ“ä½œã™ã‚‹ã®ãŒ**é‡å­ã‚²ãƒ¼ãƒˆ**ã§ã™ã€‚ã‚²ãƒ¼ãƒˆã¯æ•°å­¦çš„ã«ã¯**ãƒ¦ãƒ‹ã‚¿ãƒªè¡Œåˆ—**ã§è¡¨ç¾ã•ã‚Œã¾ã™ã€‚çŠ¶æ…‹ãƒ™ã‚¯ãƒˆãƒ«ã«ã‚²ãƒ¼ãƒˆã‚’ä½œç”¨ã•ã›ã‚‹ã“ã¨ã¯ã€ãƒ™ã‚¯ãƒˆãƒ«ã«è¡Œåˆ—ã‚’æ›ã‘ã‚‹ï¼ˆè¡Œåˆ—ç©ï¼‰ã“ã¨ã«ç›¸å½“ã—ã¾ã™ã€‚

  - **Xã‚²ãƒ¼ãƒˆ (Pauli-X / NOTã‚²ãƒ¼ãƒˆ)**: ãƒ“ãƒƒãƒˆã‚’åè»¢ã•ã›ã¾ã™ ($|0\rangle \leftrightarrow |1\rangle$)ã€‚
    $$X = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$$

  - **Hã‚²ãƒ¼ãƒˆ (ã‚¢ãƒ€ãƒãƒ¼ãƒ«ã‚²ãƒ¼ãƒˆ)**: 0ã¨1ãŒå‡ç­‰ã«æ··ã–ã£ãŸ**é‡ã­åˆã‚ã›çŠ¶æ…‹**ã‚’ä½œã‚Šå‡ºã—ã¾ã™ã€‚
    $$H = \frac{1}{\sqrt{2}}\begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix}$$

ã“ã‚Œã‚‰ã®ã‚²ãƒ¼ãƒˆã‚’è¡Œåˆ—ã¨ã—ã¦ã‚³ãƒ¼ãƒ‰ã§å®šç¾©ã—ã¾ã™ã€‚

```python
# Xã‚²ãƒ¼ãƒˆ
gate_x = np.array([
    [0, 1],
    [1, 0]
])

# Hã‚²ãƒ¼ãƒˆ
gate_h = np.array([
    [1, 1],
    [1, -1]
]) / np.sqrt(2) # 1/âˆš2 ã§å…¨ä½“ã‚’å‰²ã‚‹

print("--- X Gate ---")
print(gate_x)
print("\n--- H Gate ---")
print(gate_h)
```

## 4\. ã‚²ãƒ¼ãƒˆæ“ä½œã®ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ ğŸš€

æº–å‚™ãŒæ•´ã„ã¾ã—ãŸã€‚å®Ÿéš›ã«çŠ¶æ…‹ãƒ™ã‚¯ãƒˆãƒ«ã«ã‚²ãƒ¼ãƒˆã‚’ä½œç”¨ã•ã›ã¦ã€çŠ¶æ…‹ãŒã©ã®ã‚ˆã†ã«å¤‰åŒ–ã™ã‚‹ã‹ã‚’è¦‹ã¦ã¿ã¾ã—ã‚‡ã†ã€‚è¡Œåˆ—ã¨ãƒ™ã‚¯ãƒˆãƒ«ã®ç©ã¯ `np.dot()` é–¢æ•°ã§è¨ˆç®—ã§ãã¾ã™ã€‚

```python
# --- ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³å®Ÿè¡Œ ---

print(f"åˆæœŸçŠ¶æ…‹ |0>: {qbit_0}")
print("-" * 30)

# ä¾‹1: |0> ã« Xã‚²ãƒ¼ãƒˆ ã‚’é©ç”¨
# çµæœã¯ |1> ã«ãªã‚‹ã¯ãš
state_after_x = np.dot(gate_x, qbit_0)
print(f"X|0> ã®çµæœ: {state_after_x}")
print(f"ã“ã‚Œã¯ |1> çŠ¶æ…‹ã¨ä¸€è‡´ã—ã¾ã™ã‹ï¼Ÿ: {np.allclose(state_after_x, qbit_1)}")
print("-" * 30)


# ä¾‹2: |0> ã« Hã‚²ãƒ¼ãƒˆ ã‚’é©ç”¨
# çµæœã¯ |+> ã¨å‘¼ã°ã‚Œã‚‹é‡ã­åˆã‚ã›çŠ¶æ…‹ã«ãªã‚‹
state_after_h = np.dot(gate_h, qbit_0)
print(f"H|0> ã®çµæœ: {state_after_h}")
print("æŒ¯å¹…ãŒ [1/âˆš2, 1/âˆš2] ã¨ãªã£ã¦ã„ã¾ã™ã€‚")
print("-" * 30)


# ä¾‹3: ä¾‹2ã®çŠ¶æ…‹ã«ã‚‚ã†ä¸€åº¦ Hã‚²ãƒ¼ãƒˆ ã‚’é©ç”¨
# Hã‚²ãƒ¼ãƒˆã¯2å›é©ç”¨ã™ã‚‹ã¨å…ƒã«æˆ»ã‚‹æ€§è³ªãŒã‚ã‚Šã¾ã™ (H*H = I)ã€‚
# çµæœã¯ |0> ã«æˆ»ã‚‹ã¯ãš
state_after_hh = np.dot(gate_h, state_after_h)
print(f"H(H|0>) ã®çµæœ: {state_after_hh}")
print(f"ã“ã‚Œã¯ |0> çŠ¶æ…‹ã¨ä¸€è‡´ã—ã¾ã™ã‹ï¼Ÿ: {np.allclose(state_after_hh, qbit_0)}")
```
